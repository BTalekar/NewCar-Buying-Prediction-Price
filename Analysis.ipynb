```python
df = pd.read_csv('cars.csv')
print(df)
```

```python
dataset = df[['Manufacturer','Ex-Showroom Price(Rs)','Body_Type','Type','Fuel_Type','Drivetrain','Seating_Capacity']]
dataset.head()
```

```python
Fuel_Type=dataset[['Fuel_Type']]
Fuel_Type=pd.get_dummies(Fuel_Type, drop_first=True)

Type=dataset[['Type']]
Type=pd.get_dummies(Type, drop_first=True)

Body_Type =dataset[['Body_Type']]
Body_Type=pd.get_dummies(Body_Type, drop_first=True)

Drivetrain =dataset[['Drivetrain']]
Drivetrain=pd.get_dummies(Drivetrain, drop_first=True)

Manufacturer =dataset[['Manufacturer']]
Manufacturer=pd.get_dummies(Manufacturer, drop_first=True)


dataset=pd.concat([dataset,Fuel_Type, Type, Body_Type,Drivetrain,Manufacturer], axis=1)

dataset.drop(labels=['Fuel_Type', 'Type', 'Body_Type','Drivetrain','Manufacturer'], axis=1, inplace=True)

dataset.head()
```

```python
dataset.info()
```

```python
sell=dataset['Ex-Showroom Price(Rs)']
dataset.drop(['Ex-Showroom Price(Rs)'], axis=1, inplace=True)
dataset=dataset.join(sell)
dataset.head()
```

```python
X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor
dt_reg = DecisionTreeRegressor(random_state = 0)
dt_reg.fit(X_train, y_train)
y_pred=dt_reg.predict(X_test)

print("Decision Tree Score on Training set is",dt_reg.score(X_train, y_train))#Training Accuracy
print("Decision Tree Score on Test Set is",dt_reg.score(X_test, y_test))#Testing Accuracy

accuracies = cross_val_score(dt_reg, X_train, y_train, cv = 5)
print(accuracies)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
```

```python
#Random Forest Regression
from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor(n_estimators=400,min_samples_split=15,min_samples_leaf=2,
max_features='auto', max_depth=30)
rf_reg.fit(X_train, y_train)
y_pred=rf_reg.predict(X_test)

print("Random Forest Score on Training set is",rf_reg.score(X_train, y_train))#Training Accuracy
print("Random Forest Score on Test Set is",rf_reg.score(X_test, y_test))#Testing Accuracy

accuracies = cross_val_score(rf_reg, X_train, y_train, cv = 5)
print(accuracies)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
```
